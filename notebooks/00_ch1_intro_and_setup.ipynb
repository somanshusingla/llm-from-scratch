## What is a Large Language Model (LLM)?

An LLM is a neural network trained on enormous amounts of text with one deceptively
simple objective: **predict the next word (token)**. Do that well enough, at
scale, and the model picks up grammar, facts, reasoning patterns, translation, and
more — all as a side effect of getting good at "what comes next?".

- **"Large"** = billions of parameters (adjustable numbers) trained on
  hundreds of billions of words. The model we build is a small GPT-2 (124M
  parameters) — the same *architecture* as models behind modern chatbots, just
  smaller.
- **"Language Model"** = a model that assigns probabilities to sequences of
  words and can generate new text.

## What can they do?

Text generation, summarization, translation, question answering, sentiment
analysis, code generation, chat assistants — all from that single next-token
objective, sometimes with a little task-specific **fine-tuning** (chapters 6–7).

## The two stages of building an LLM

1. **Pretraining** — train on a massive, *unlabeled* text corpus to predict the
   next token. This is expensive and produces a general-purpose "base" model.
   *(Chapters 2–5.)*
2. **Fine-tuning** — take the pretrained base model and adapt it with a smaller,
   *labeled* dataset to a specific job:
   - **Classification fine-tuning** — e.g. spam vs. not-spam *(Chapter 6)*.
   - **Instruction fine-tuning** — teach it to follow instructions / chat
     *(Chapter 7)*.

We can't afford to pretrain a giant model ourselves, so in Chapter 5 we load
**OpenAI's pretrained GPT-2 weights** into the model we build — same architecture,
real knowledge.

## The transformer & GPT architecture

Modern LLMs are built on the **transformer** (2017). The original transformer had
two halves — an *encoder* (understands input) and a *decoder* (produces output),
designed for translation.

**GPT uses only the decoder half.** Its superpower is the **attention mechanism**:
when processing a word, the model can "look at" every other relevant word in the
context and weigh how much each matters. That's what lets it resolve "it" to the
right noun, track a topic across a paragraph, and stay coherent.

At a glance, the GPT you'll build is:

```
tokens -> [token embeddings + positional embeddings]
       -> [ transformer block ] x 12   (each = multi-head attention + feed-forward)
       -> [ final layer norm ]
       -> [ output head ] -> probabilities over the next token
```

Every one of those boxes is something you implement, understand, and run in the
next six notebooks.

````python
# A 30-second taste of what's coming: turn text into the numbers an LLM sees.
tokenizer = tiktoken.get_encoding("gpt2")   # the GPT-2 tokenizer
text = "Large language models predict the next token."
ids = tokenizer.encode(text)
print("Text  :", text)
print("Tokens:", ids)
print("Back  :", tokenizer.decode(ids))
print("\nThat mapping text <-> numbers is exactly where Chapter 2 begins.")
````
